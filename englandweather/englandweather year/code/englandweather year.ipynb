```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("englandweather.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df2 = pd.DataFrame(data, columns=['Summary', 'Precip Type'])
df2
```

```python
df2.describe()
```

```python
df3 = df.dropna()
df4 = df2.dropna()
```

```python
df3
```

```python

df5 = df3['Formatted Date'].str.split(' ', expand=True)
df3['Formatted Date'] = df5[0]
df6 = df3.rename({'Formatted Date':'Date'}, axis=1)
df6

```

```python
Year = pd.to_datetime(df6.Date).dt.year
```

```python
plt.figure(figsize=(10,6), dpi=800)
plt.scatter(df3['Temperature (C)'], Year)
plt.xlabel('Temperature (C)', fontsize=15)
plt.ylabel('Year', fontsize=15)
plt.title('Annual temperature changes in England', fontsize=20)
plt.grid()
plt.savefig('temp-year.jpg', bbox_inches='tight', pad_inches=1)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=800)
plt.scatter(df3['Wind Speed (km/h)'], Year)
plt.xlabel('Wind Speed (km/h)', fontsize=15)
plt.ylabel('Year', fontsize=15)
plt.title('Annual wind speed variations in England', fontsize=20)
plt.grid()
plt.savefig('wind-year.jpg', bbox_inches='tight', pad_inches=1)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=800)
plt.scatter(df3['Humidity'], Year)
plt.xlabel('Humidity', fontsize=15)
plt.ylabel('Year', fontsize=15)
plt.title('Annual humidity variations in England', fontsize=20)
plt.grid()
plt.savefig('Humidity-year.jpg', bbox_inches='tight', pad_inches=1)
plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=800)
plt.scatter(df3['Pressure (millibars)'], Year)
plt.xlabel('Pressure (millibars)', fontsize=15)
plt.ylabel('Year', fontsize=15)
plt.title('Annual pressure changes in England', fontsize=20)
plt.grid()
plt.savefig('Pressure (millibars)-year.jpg', bbox_inches='tight', pad_inches=1)
plt.show()
```
