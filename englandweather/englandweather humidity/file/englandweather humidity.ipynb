```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("englandweather.csv")
data
```

```python
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df2 = pd.DataFrame(data, columns=['Summary', 'Precip Type'])
df2
```

```python
df2.describe()
```

```python
df3 = df.dropna()
df4 = df2.dropna()
```

```python
df3
```

```python
df4
```

```python
plt.figure(figsize=(10, 7), dpi=1200)
plt.scatter(df3['Temperature (C)'], df3['Humidity'])
plt.xlabel('Temperature', fontsize=20, color='green')
plt.ylabel('Humidity', fontsize=20, color='green')
# plt.title(label="ReLU function graph",
#           fontsize=40,
#           color="green")
plt.title(label="The effect of temperature on humidity in England",
          fontsize=20,
          color='red')
#plt.yticks(rotation=360) # baraye zamani ke neveshteha roye ham mioftan karbord dare
plt.grid()
plt.savefig('Temp-Humidity.jpg',bbox_inches='tight',pad_inches= 1)
plt.show()
```

```python
plt.figure(figsize=(10, 7), dpi=1200)
plt.scatter(df3['Pressure (millibars)'], df3['Humidity'])
plt.xlabel('Pressure (millibars)', fontsize=20, color='green')
plt.ylabel('Humidity', fontsize=20, color='green')
plt.title(label="The effect of Pressure on humidity in England",
           fontsize=20,
           color='red')
#plt.yticks(rotation=360) # baraye zamani ke neveshteha roye ham mioftan karbord dare
plt.grid()
plt.savefig('Pressure-humidity.jpg',bbox_inches='tight',pad_inches= 1)
plt.show()
```

```python
plt.figure(figsize=(10, 7), dpi=1200)
plt.scatter(df3['Wind Speed (km/h)'], df3['Humidity'])
plt.xlabel('Wind Speed (km/h)', fontsize=20, color='green')
plt.ylabel('Humidity', fontsize=20, color='green')
plt.title(label="The effect of Wind Speed on humidity in England",
           fontsize=20,
           color='red')
#plt.yticks(rotation=360) # baraye zamani ke neveshteha roye ham mioftan karbord dare
plt.grid()
plt.savefig('Wind_Speed-humidity.jpg',bbox_inches='tight',pad_inches= 1)
plt.show()
```
